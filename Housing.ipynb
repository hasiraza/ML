```python


# Set style for better aesthetics
sns.set(style="whitegrid")

# Create figure with proper size
plt.figure(figsize=(8, 4))

# Boxplot
sns.boxplot(
    x=df['total_rooms'],
    color='skyblue',
    linewidth=1.5
)

# Add title and labels
plt.title('Distribution of Total rooms', fontsize=14, weight='bold')
plt.xlabel('Housing Median Age', fontsize=12)

# Improve layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Data
original = df['total_rooms']
trans_log = np.log1p(original)
trans_squre = np.square(trans_log)
trans_cube = np.cbrt(trans_log)

# Figure (2 rows, 2 columns)
plt.figure(figsize=(12, 10))

# 🔹 Original
plt.subplot(2, 2, 1)
sns.histplot(original, kde=True, bins=30, color='steelblue')
plt.title(f'Original (Skew: {original.skew():.2f})')

# 🔹 Log1p
plt.subplot(2, 2, 2)
sns.histplot(trans_log, kde=True, bins=30, color='orange')
plt.title(f'Log1p (Skew: {trans_log.skew():.2f})')

# 🔹 Sqrt(Log1p)
plt.subplot(2, 2, 3)
sns.histplot(trans_squre, kde=True, bins=30, color='green')
plt.title(f'Squre(Log1p) (Skew: {trans_squre.skew():.2f})')

# 🔹 Cube root(Log1p)
plt.subplot(2, 2, 4)
sns.histplot(trans_cube, kde=True, bins=30, color='purple')
plt.title(f'Cube Root(Log1p) (Skew: {trans_cube.skew():.2f})')

plt.tight_layout()
plt.show()

# Print skew values
print("Original Skewness:", original.skew())
print("Log1p Skewness:", trans_log.skew())
print("Squre(Log1p) Skewness:", trans_squre.skew())
print("Cbrt(Log1p) Skewness:", trans_cube.skew())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X=data.drop('median_house_value',axis=1)
y=data['median_house_value']
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.drop('median_house_value',axis=1),data['median_house_value'],test_size=0.2,random_state=42)
```

```python
X_train_scaled=StandardScaler().fit_transform(X_train)
X_test_scaled=StandardScaler().fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA
pc=PCA(n_components=8)
X_train=pc.fit_transform(X_train_scaled)
X_test=pc.transform(X_test_scaled)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
r2_score(y_test,y_predict)
```

```python
from sklearn.linear_model import SGDRegressor
model=SGDRegressor(loss='squared_error', max_iter=1000, alpha=0.1)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
r2_score(y_test,y_predict)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# predictions
y_pred = model.predict(X_test)

plt.figure(figsize=(6,6))

sns.scatterplot(x=y_test, y=y_pred)

# perfect linear line
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linewidth=2)

plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted (Linear Fit Check)")
plt.savefig("actual_vs_predicted.png")
plt.show()
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
